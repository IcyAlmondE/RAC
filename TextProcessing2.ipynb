```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

service = Service("C:\\Users\\Asus\\OneDrive\\Desktop\\IcyAlmondE\\RAC\\16-25 Oct\\chromedriver-win64\\chromedriver.exe")
driver = webdriver.Chrome(service=service)
# driver = webdriver.Chrome('/Users/sukreesinthupinyo/App/chromedriver')
driver.implicitly_wait(30)
driver.maximize_window()
```

```python
driver.get('https://www.thairath.co.th/home')
```

```python
from bs4 import BeautifulSoup

soup=BeautifulSoup(driver.page_source, 'html.parser')
tab_content=soup.find('div', {'class':'tab-content'})
```

```python
recs=[]
for d in tab_content.find_all('div', {'class':'tab-pane'}):
    atts = d.find_all('a')
    for att in atts:
        print(att.text)
        rec={'url':att['href'], 'news':att['title']}
        recs.append(rec)

recs
```

```python
import pandas as pd
import time

df=pd.DataFrame(recs)
df
```
