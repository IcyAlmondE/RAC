# Regular Expression

```python
import re

# Sample text containing email addresses
text = "Please contact support@example.com for assistance or info@company.com for more information."

# Define a regular expression pattern for matching email addresses
email_pattern = r'\b[\w.-]+@[\w.-]+\.\w+\b'

# Use the re.findall() function to find all email addresses in the text
email_addresses = re.findall(email_pattern, text)

# Print the extracted email addresses
for email in email_addresses:
    print(email)
```
